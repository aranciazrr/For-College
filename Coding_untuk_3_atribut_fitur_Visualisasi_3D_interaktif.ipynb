```python
import pandas as pd
from sklearn.cluster import KMeans
import plotly.graph_objects as go

data = {
    'ID Pelanggan': [1, 2, 3, 4, 5, 6, 7],
    'Umur': [22, 35, 47, 25, 60, 33, 40],
    'Pendapatan': [3000, 5000, 8000, 3500, 12000, 4500, 5500],
    'Jumlah Pembelian': [200, 400, 600, 250, 700, 350, 450]
}
df = pd.DataFrame(data)


X = df[['Umur', 'Pendapatan', 'Jumlah Pembelian']]


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)


fig = go.Figure()

for cluster in df['Cluster'].unique():
    cluster_data = df[df['Cluster'] == cluster]
    fig.add_trace(go.Scatter3d(
        x=cluster_data['Umur'],
        y=cluster_data['Pendapatan'],
        z=cluster_data['Jumlah Pembelian'],
        mode='markers',
        marker=dict(size=8),
        name=f'Cluster {cluster}'
    ))

fig.update_layout(
    title='Segmentasi Pelanggan 3D Interaktif',
    scene=dict(
        xaxis_title='Umur',
        yaxis_title='Pendapatan',
        zaxis_title='Jumlah Pembelian'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```
