#Input Library python

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('data_produk.csv')
print(dataset)
```

# Deteksi Value yang hilang

```python
dataset.isna().values.any() #bisa pakai isnull
```

```python
print(dataset.isna().sum())
```

#pengisian data yang hilang

```python
dataset.at[7, 'Produk'] = 'Produk A'
print(dataset)
```

Pengisian data yang hilang ke - 2


```python
dataset.at[9, 'Pendapatan'] = 12 * 150
print(dataset)
```

# menghapus data duplikat

```python
dataset[dataset.duplicated()]
```

```python
dataset.drop(14, inplace=True)
print(dataset)
```

#menambahkan harga satuan

Harga Hatuan = Pendapatan / Jumlah Terjual

```python
dataset['Harga Satuan'] = dataset['Pendapatan'] / dataset['Jumlah Terjual']
print(dataset)
```

# Visualisasi data

line chart
menggunakan seaborn python

```python
dataset.dtypes
```

```python
import seaborn as sns

#konversi tanggal ke datetime
dataset['Tanggal'] = pd.to_datetime(dataset['Tanggal'])
dataset.dtypes
```

# Menstandarkan nama lokasi

```python
dataset.at[7, 'Lokasi'] = 'Jakarta'
dataset.at[9, 'Lokasi'] = 'Surabaya'
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=dataset, x='Tanggal', y='Jumlah Terjual', hue='Produk', marker='o')
plt.title('Tren Penjualan Produk dari waktu ke waktu')
plt.xlabel('Tanggal Penjualan')
plt.ylabel('Jumlah Terjual')
plt.legend(title='Produk')
plt.grid(True)
plt.show
```

Pie Chart

```python
plt.figure(figsize=(8, 8))
dataset['Lokasi'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribusi penjualan berdasarkan Lokasi')
plt.show()
```
